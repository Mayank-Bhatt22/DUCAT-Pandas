# 6. **Indexing and Selection**
| Method            | Description                          |
|-------------------|--------------------------------------|
| `s[0]`, `s['a']`  | Access by position or label          |
| `s.loc['a']`      | Label-based selection                |
| `s.iloc[0]`       | Position-based selection             |
| `s.at['a']`       | Fast label access                    |
| `s.iat[0]`        | Fast position access                 |
| `s.get(key)`      | Safe getter                          |

---
- **loc:**
It selects data by labels (row and column names). It can handle single labels, lists of labels, and slices with labels.
- **iloc:**
It selects data by integer positions (row and column indices). It also accepts single integers, lists of integers, and slices with integers.
- **at:**
It accesses a single value by label. It is optimized for speed when accessing individual elements.
- **iat:**
It accesses a single value by integer position. Similar to at, it is optimized for fast access to scalar values.
